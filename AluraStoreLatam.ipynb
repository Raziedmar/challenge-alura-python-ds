###Importaci√≥n de datos

```python
from IPython.display import display, HTML , clear_output
import ipywidgets as widgets
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tienda2.head()
tienda3.head()
tienda4.head()
```

###Exploraci√≥n de datos

Se realizar√° una exploraci√≥n de los datos de los datos de las 4 datas disponibles. Utilizando los m√©todos:

*   .*info()*
*   .*describe()*
*   .*value_conts()*

 y el atributo *shape*


```python
# Crea el diccionario que relaciona una clave de tienda con su respectivo valor de variable de dataframe segun la tienda
opciones={'Tienda 1':tienda,'Tienda 2':tienda2, 'Tienda 3':tienda3,'Tienda 4':tienda4}

def mostrar_info(change):
    # Esto ayuda a limpiar la celda de salida cada vez que cambias la selecci√≥n
    from IPython.display import clear_output
    clear_output(wait=True)

    # Muestra el desplegable
    display(w_opciones)

    # Captura el valor seleccionado
    df_seleccionado = opciones[change['new']]
    nombre_tienda = change['new']

    # --- 1. Muestra .info() (Impresi√≥n directa) ---
    display(HTML(f"<h3><b>--- üìä Informaci√≥n Estructural de {nombre_tienda} ---</b></h3>"))
    df_seleccionado.info()

    # --- 2. Muestra .describe() (Visualizaci√≥n de tabla) ---
    display(HTML(f"<h3><b>--- üìà Estad√≠stica Descriptiva de {nombre_tienda} ---</b></h3>"))
    display(df_seleccionado.describe())

    # --- 3. Muestra el Shape (Dimensiones) ---
    filas = df_seleccionado.shape[0]
    columnas = df_seleccionado.shape[1]
    print('\n')
    display(HTML(f'<b>--- üî¢ La BD {nombre_tienda} tiene {filas} filas y {columnas} columnas.</b>'))

    # --- 4. Muestra .value_consts() ---
    display(HTML(f"<h3><b>--- üìà Estad√≠stica Descriptiva de {nombre_tienda} ---</b></h3>"))
    display(df_seleccionado['Lugar de Compra'].value_counts())

# Crea los widget Dropdown
w_opciones = widgets.Dropdown(
    options=opciones.keys(),
    value='Tienda 1',  # Valor predeterminado
    description='Selecciona BD de Tienda:',
    disabled=False,
)

# Observa el cambio del valor seleccionado en el desplegable y llama a la funci√≥n que muestra los reportes de los metodos de exploraci√≥n de datos
w_opciones.observe(mostrar_info, names='value')
# Ejecuta la funci√≥n una vez con el valor predeterminado para ver el reporte inicial
mostrar_info({'new': w_opciones.value})
```
